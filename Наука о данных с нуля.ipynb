```python
movies = ["Энни Холл", "Бен-Гур", "Касабланка", "Ганди", "Вестсайдская \n история"]
num_oscars = [5, 11, 3, 8, 10]
# Построить столбцы с левыми Х-координатами [xs] и высотами [num_oscars]
plt.bar(range(len(movies)), num_oscars)
plt.title("Moи любимые фильмы")
plt.ylabel("Количество наград")
# Добавить заголовок
# Разместить метку на оси у
# Пометить ось х названиями фильмов в центре каждого столбца
plt.xticks(range(len(movies)), movies)
plt.show()
```

# Линейный графики

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
# Квадрат смещения
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
# Суммарная ошибка
total_error = [х + у for х, у in zip(variance, bias_squared)]
xs = [i for i,_ in enumerate(variance)]
                
# Метод plt.plot можно вызывать много раз,
# чтобы показать несколько графиков на одной и той же диаграмме:

# зеленая сплошная линия
plt.plot(xs, variance,'g-', label='дисперсия')
# красная штрихпунктирная
plt.plot(xs, bias_squared, 'r-.', label='смещение^2')
# синяя пунктирная
plt.plot(xs, total_error, 'b:', label='суммарная ошибка')

# Ее.ли для каждой линии задано название label,
# то легенда будет показана автоматически,
# loc=9 означает "наверху посередине"
plt.legend(loc=9)
plt.xlabel("Сложность модели")
plt.title("Koмпpoмиcc между смещением и дисперсией")
plt.show()
```

# Диаграммы рассеяния

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['а', 'b', 'с', 'd', 'е', 'f', 'g', 'h', 'i']
# Друзья
# Минуты
# Метки
plt.scatter(friends, minutes)

# Назначить метку для каждой точки
for label, friend_count, minute_count in zip(labels, friends, minutes):
    # Задать метку и немного сместить ее
    plt.annotate(label, xy=(friend_count, minute_count), xytext=(5, -5),textcoords='offset points')
    
plt.title("Чиcлo минут против числа друзей")
plt.xlabel("Число друзей")
plt.ylabel("Чиcлo минут, проводимых на сайте ежедневно")
plt.show()
```

```python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]
# Оценки за экзамен 1
# Оценки за экзамен 2

plt.scatter(test_1_grades, test_2_grades)
plt.title("Ocи не сопоставимы")
plt.xlabel("Оценки за экзамен 1")
plt.ylabel("Оценки за экзамен 2")
plt.axis('equal') #  чтобы оси были соразмерны
plt.show()
```
